```python
#selecting biopsy data as target values

target = df['Biopsy']
input = df.drop(['Biopsy'], axis=1)
```

```python
X = np.array(input).astype('float32')
y = np.array(target).astype('float32')

y = y.reshape(-1,1)
```

```python
scaler  = StandardScaler()
X = scaler.fit_transform(X)

X
```

```python
#splitting into training and test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5)
```

```python
#fitting data into model


model = xgb.XGBClassifier(learning_rate = 0.1, max_depth = 50, n_estimators = 100)

model.fit(X_train, y_train)
```

```python
#testing results of our model
result_test = model.score(X_test, y_test)

result_test
```

```python
y_predict = model.predict(X_test)
```

```python

print(classification_report(y_test, y_predict))
```

```python
cm = confusion_matrix(y_predict, y_test)

sns.heatmap(cm, annot=True)

plt.show()
```
